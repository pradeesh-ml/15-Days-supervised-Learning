```python
#import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.utils import resample
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from xgboost import XGBClassifier
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
```

```python
# Load the dataset
df = pd.read_csv('creditcard.csv')
df.head()
```

```python
# Data visualization
sns.countplot(x='Class', data=df)
plt.title('Class Distribution')
plt.show()
```

```python
# Visualize correlations
plt.figure(figsize=(20, 16))
sns.heatmap(df.corr(), annot=False, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Resampling for class imbalance
majority_class = df[df['Class'] == 0]
minority_class = df[df['Class'] == 1]
# Downsample the majority class
majority_downsampled = resample(majority_class, 
                                 replace=True, 
                                 n_samples=int(len(minority_class) * 5), 
                                 random_state=101)
```

```python
# Combine minority class with downsampled majority class
df_downsampled = pd.concat([minority_class, majority_downsampled]).sample(frac=1, random_state=101).reset_index(drop=True)
```

```python
# Split the data into features and target
X = df_downsampled.drop('Class', axis=1)
y = df_downsampled['Class']
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101, stratify=y)
```

```python
# Feature scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Model training
xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
xgb.fit(X_train, y_train)
```

```python
# Predictions
train_prediction = xgb.predict(X_train)
test_prediction = xgb.predict(X_test)
```

```python
# Evaluation
print("Training Set Evaluation")
print(confusion_matrix(y_train, train_prediction))
print(classification_report(y_train, train_prediction))
```

```python
print("Test Set Evaluation")
print(confusion_matrix(y_test, test_prediction))
print(classification_report(y_test, test_prediction))
```

```python
# Confusion matrix visualization
cm = confusion_matrix(y_test, test_prediction)
display = ConfusionMatrixDisplay(cm)
display.plot(cmap='Blues')
plt.title('Confusion Matrix for Test Set')
plt.show()
```
